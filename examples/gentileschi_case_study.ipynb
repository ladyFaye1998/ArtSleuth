# The Gentileschi Workshop — A Case Study

The question of attribution within the Gentileschi workshop
has occupied art historians for over a century. When Orazio
and Artemisia Gentileschi collaborated — father and daughter,
both formidable painters — their individual contributions to
a shared canvas can be fiendishly difficult to disentangle.

Artemisia's characteristic "action nodes" — the compositional
inflection points where narrative tension peaks — carry a
distinctive brushstroke signature: higher coherence, steeper
orientation gradients, and greater impasto energy than the
surrounding passages.

This notebook demonstrates how ArtSleuth's brushstroke
clustering can reveal these signatures computationally,
providing quantitative support for the qualitative judgements
of traditional connoisseurship.

---

## Methodology

1. Load a Gentileschi workshop painting.
2. Run brushstroke analysis with adaptive patch extraction.
3. Visualise cluster assignments overlaid on the painting.
4. Compare stroke statistics across clusters to identify
   passages attributable to distinct hands.

```python
from PIL import Image
from artsleuth.config import AnalysisConfig
from artsleuth.core.brushstroke import BrushstrokeAnalyzer
from artsleuth.utils.visualization import plot_brushstroke_map

config = AnalysisConfig(
    patch_size=128,
    patch_strategy="adaptive",
    device="cpu",
)

image = Image.open("gentileschi_judith.jpg").convert("RGB")
analyzer = BrushstrokeAnalyzer(config)
report = analyzer.analyze(image)

print(f"Patches analysed: {len(report.descriptors)}")
print(f"Distinct clusters: {len(set(report.cluster_labels.tolist()))}")
print(f"Mean coherence: {report.mean_coherence:.3f}")
```

```python
# Visualise brushstroke clusters on the painting surface
plot_brushstroke_map(report, image, "gentileschi_clusters.png")

Image.open("gentileschi_clusters.png")
```

```python
import numpy as np

# Compare stroke statistics across clusters
for cluster_id in sorted(set(report.cluster_labels.tolist())):
    mask = report.cluster_labels == cluster_id
    cluster_descs = [d for d, m in zip(report.descriptors, mask) if m]

    coherences = [d.coherence for d in cluster_descs]
    energies = [d.energy for d in cluster_descs]

    print(f"Cluster {cluster_id}: "
          f"n={len(cluster_descs)}, "
          f"coherence={np.mean(coherences):.3f} \u00b1 {np.std(coherences):.3f}, "
          f"energy={np.mean(energies):.3f} \u00b1 {np.std(energies):.3f}")
```

## Interpretation

If the workshop hypothesis holds, we expect to see:

- **Cluster A** (Artemisia's action nodes): Higher coherence,
  greater energy, concentrated around the compositional
  inflection points (Judith's arm, Holofernes' face).
- **Cluster B** (workshop passages): Lower coherence, reduced
  energy, distributed across background and drapery.

The quantitative separation between clusters provides
measurable evidence for the multi-hand hypothesis — evidence
that complements, but does not replace, traditional
connoisseurship.
