# ArtSleuth — Quick Start

This notebook walks through the core ArtSleuth workflow:
analysing a painting for style, brushstroke patterns,
attribution, and generating visual explanations.

## 1. Full Pipeline Analysis

The simplest entry point — one function call, one structured report.

```python
import artsleuth

result = artsleuth.analyze("your_painting.jpg")
print(result.summary())
```

## 2. Style Classification

Classify the painting by historical period, art-historical school,
and material technique.

```python
from PIL import Image
from artsleuth.config import AnalysisConfig
from artsleuth.core.style import StyleClassifier

config = AnalysisConfig(device="cpu")
classifier = StyleClassifier(config)

image = Image.open("your_painting.jpg").convert("RGB")
style_report = classifier.classify(image, top_k=5)

print(f"Period:    {style_report.period.label} ({style_report.period.confidence:.0%})")
print(f"School:    {style_report.school.label} ({style_report.school.confidence:.0%})")
print(f"Technique: {style_report.technique.label} ({style_report.technique.confidence:.0%})")
```

## 3. Visual Explanation

Generate a Grad-CAM heatmap showing which regions of the painting
most influenced the model's verdict.

```python
explanation = result.explain(target="style")
explanation.save("explanation_overlay.png")

# Display inline
from PIL import Image as PILImage
PILImage.fromarray(explanation.composite)
```

## 4. Brushstroke Analysis

Examine the physical evidence of the painter's hand —
stroke orientation, coherence, and energy.

```python
from artsleuth.core.brushstroke import BrushstrokeAnalyzer

analyzer = BrushstrokeAnalyzer(config)
brush_report = analyzer.analyze(image)

print(f"Mean coherence: {brush_report.mean_coherence:.3f}")
print(f"Patches analysed: {len(brush_report.descriptors)}")
print(f"Distinct clusters: {len(set(brush_report.cluster_labels.tolist()))}")
```
